```python
import pandas as pd
import sqlite3 as sql
```

# Leitura dos dados .csv 

```python
adidos_auxiliares = pd.read_csv("bd-csv\\adidos-auxiliares.csv")
almoco_servido_EB_2017 = pd.read_csv("bd-csv\\almoco-servido-EB-2017.csv", sep=';', encoding='ANSI')
cafe_servido_EB_2017 = pd.read_csv("bd-csv\cafe-servido-EB-2017.csv", sep=';', encoding='ANSI')
cma = pd.read_csv("bd-csv\cma.csv")
jantar_servido_EB_2017 = pd.read_csv("bd-csv\\jantar-servido-EB-2017.csv", sep=';', encoding='ANSI')
jsm2020 = pd.read_csv("bd-csv\jsm2020.csv")
orgao2020 = pd.read_csv("bd-csv\orgao2020.csv")
psc2019 = pd.read_csv("bd-csv\psc2019.csv")
sermil2019 = pd.read_csv("bd-csv\sermil2019.csv")
```

# Criação de um BD SQL Lite

```python
conn = sql.connect('exercito.db')
```

## Escrita dos dados no BD

```python
adidos_auxiliares.to_sql('adidos_auxiliares', conn)
almoco_servido_EB_2017.to_sql('almoco_servido_EB_2017', conn)
cafe_servido_EB_2017.to_sql('cafe_servido_EB_2017', conn)
cma.to_sql('cma', conn)
jantar_servido_EB_2017.to_sql('jantar_servido_EB_2017', conn)
jsm2020.to_sql('jsm2020', conn)
orgao2020.to_sql('orgao2020', conn)
psc2019.to_sql('psc2019', conn)
sermil2019.to_sql('sermil2019', conn)
```

## Teste no BD criado

```python
pd.read_sql('SELECT * FROM psc2019 p INNER JOIN cma c on p.cmd_mil_area = c.sigla', conn)
```

## Finalizando processo

```python
conn.close()
```
